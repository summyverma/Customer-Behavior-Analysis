```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
# now check again null values
df.isnull().sum()
```

```python
#we can see column nmaes contain name with spaces with capital letters
# we will change it to dash instead of spaces with small letters
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
# here we want to change the name of purchase amount usd to  purchase amount
```

```python
df=df.rename(columns={'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
df.columns
```

```python
df
```

```python
#again 
labels = ['Young_Adult','Adult','Middle_aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create column purchase_frequency_days 

frequency_mapping =  {
    'Fortnightly' : 14,
    'Weekly'      : 7,
    'Monthly' :    30,
    'Quarterly'  : 90,
    'Bi-weekly' : 14,
    'Annually'  : 365,
    'Every 3 Months' : 90 
}
df ['Purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
    
```
